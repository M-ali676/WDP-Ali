```python
cart=[]


print("Welcome to Ecommerce Website")
while True:
    select=input("Which Category you want to shop: \n1.Vegetable and fruits \n2.Electronics \n3.Stationary \n4grpcery")
    if select=="1":
        print("Welcome to veg and fruit section!")
        vf=input("What you want to add? (Veg or fruits) ?").lower()
        if vf=="veg":
            while True:
                veg=input("palease enter the name of vegetable you want to add: ")
                if veg=="done":
                    break
                cart.append(veg)
        elif vf=="fruits":
            while True:
                fruits=input("please enter the name of fruit you want to add: ")
                if fruits=="done":
                    break
                cart.append(fruits)
    elif select=="2":
        print("Welcome to electronics section!")
        elec=input("What you want to add? (Mobile or laptop) ?").lower()
        if elec=="mobile":
            while True:
                mobile=input("Please enter the name of mobile you want to add: ")
                if mobile=="done":
                    break
                cart.append(mobile)
        elif elec=="laptop":
            while True:
                laptop=input("Please enter the name of laptop you want to add: ")
                if laptop=="done":
                    break
                cart.append(laptop)
    elif select=="3":
        print("Welcome to stationary section!")
        sta=input("What you want to add? (Pen or notebook) ?").lower()
        if sta=="pen":
            while True:
                pen=input("please enter the name of pen you want to add: ")
                if pen=="done":
                    break
                cart.append(pen)
        elif sta=="notebook":
            while True:
                notebook=input("nplease enter the name of notebook you want to add: ")
                if notebook=="done":
                    break
                cart.append(notebook)
    elif select=="4":
        print("Welcome to grocery section!")
        groc=input("What you want to add? (Rice or dal) ?").lower()
        if groc=="rice":
            while True:
                rice=input("Please enter the name of rice you want to add: ")
                if rice=="done":
                    break
                cart.append(rice)
        elif groc=="dal":
            while True:
                dal=input("Please enter the name of dal you want to add: ")
                if dal=="done":
                    break
                cart.append(dal)
    else:
        print("Invalid selection! Please select a valid category.")
    continue_shopping=input("Do you want to continue shopping? (yes/no)").lower()
    if continue_shopping!="yes":
        break
print("Your cart items are:")
for item in cart:
    print(item)
```
